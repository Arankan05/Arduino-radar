```python
import pandas as pd
import matplotlib.pyplot as plt

# Display all columns
pd.set_option('display.max_columns', None)

print("Libraries imported successfully!")
```

```python
df = pd.read_csv("Amazon.csv")

print("Dataset loaded successfully!")
```

```python
df.head()
```

```python
print("Rows and Columns:", df.shape)
```

```python
print("Column Names:")
print(df.columns)
```

```python
print("Data Types:")
print(df.dtypes)
```

```python
df.info()
```

```python
df.describe()
```

```python
print("Missing Values:")
print(df.isnull().sum())
```

```python
df["OrderDate"] = pd.to_datetime(df["OrderDate"])
```

```python
df["Profit"] = df["TotalAmount"] * 0.20
```

```python
df.head()
```

```python
total_sales = df["TotalAmount"].sum()
print(f"Total Sales: {total_sales:,.2f}")
```

```python
total_profit = df["Profit"].sum()
print(f"Total Profit: {total_profit:,.2f}")
```

```python
total_orders = df["OrderID"].nunique()
print(f"Total Orders: {total_orders}")
```

```python
average_sales = df["TotalAmount"].mean()
print(f"Average Sales: {average_sales:,.2f}")
```

```python
average_profit = df["Profit"].mean()
print(f"Average Profit: {average_profit:,.2f}")
```

```python
max_sale = df["TotalAmount"].max()
print(f"Maximum Sale: {max_sale:,.2f}")
```

```python
min_sale = df["TotalAmount"].min()
print(f"Minimum Sale: {min_sale:,.2f}")
```

```python
print("*" * 50)
print("             AMAZON SALES KPI SUMMARY")
print("*" * 50)
print(f"Total Sales      : {total_sales:,.2f}")
print(f"Total Profit     : {total_profit:,.2f}")
print(f"Total Orders     : {total_orders}")
print(f"Average Sales    : {average_sales:,.2f}")
print(f"Average Profit   : {average_profit:,.2f}")
print(f"Maximum Sale     : {max_sale:,.2f}")
print(f"Minimum Sale     : {min_sale:,.2f}")
print("*" * 50)
```

```python
# Total Sales by State

sales_by_state = (
    df.groupby("State")["TotalAmount"]
      .sum()
      .sort_values(ascending=False)
)

sales_by_state

print("========== SALES BY STATE ==========\n")
print(sales_by_state)
```

```python
sales_by_category = (
    df.groupby("Category")["TotalAmount"]
      .sum()
      .sort_values(ascending=False)
)

sales_by_category

print("========== SALES BY CATEGORY ==========\n")
print(sales_by_category)
```

```python
sales_by_brand = (
    df.groupby("Brand")["TotalAmount"]
      .sum()
      .sort_values(ascending=False)
)

sales_by_brand

print("========== SALES BY BRAND ==========\n")
print(sales_by_brand)
```

```python
highest_state = sales_by_state.idxmax()
highest_sales = sales_by_state.max()

print(f"Highest Sales State : {highest_state}")
print(f"Sales : {highest_sales:,.2f}")
```

```python
lowest_state = sales_by_state.idxmin()
lowest_sales = sales_by_state.min()

print(f"Lowest Sales State : {lowest_state}")
print(f"Sales : {lowest_sales:,.2f}")
```

```python
highest_category = sales_by_category.idxmax()
highest_category_sales = sales_by_category.max()

print(f"Highest Revenue Category : {highest_category}")
print(f"Revenue : {highest_category_sales:,.2f}")
```

```python
highest_brand = sales_by_brand.idxmax()
highest_brand_sales = sales_by_brand.max()

print(f"Highest Revenue Brand : {highest_brand}")
print(f"Revenue : {highest_brand_sales:,.2f}")
```

```python
# Top 5 Customers by Sales

top5_customers = (
    df.groupby("CustomerName")["TotalAmount"]
      .sum()
      .sort_values(ascending=False)
      .head(5)
)

print("===== TOP 5 CUSTOMERS =====")
print(top5_customers)
```

```python
product_quantity = (
    df.groupby("ProductName")["Quantity"]
      .sum()
      .sort_values(ascending=False)
)

print("===== PRODUCT QUANTITY SOLD =====")
print(product_quantity)
```

```python
top10_products = (
    df.groupby("ProductName")["TotalAmount"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

print("===== TOP 10 PRODUCTS =====")
print(top10_products)
```

```python
payment_mode = (
    df.groupby("PaymentMethod")["OrderID"]
      .count()
      .sort_values(ascending=False)
)

print(payment_mode)
```

```python
df["Month"] = df["OrderDate"].dt.strftime("%B")

month_order = [
    "January","February","March","April","May","June",
    "July","August","September","October","November","December"
]

df["Month"] = pd.Categorical(
    df["Month"],
    categories=month_order,
    ordered=True
)

monthly_sales = (
    df.groupby("Month")["TotalAmount"]
      .sum()
      .sort_index()
)

print(monthly_sales)
```

```python
highest_month = monthly_sales.idxmax()
highest_month_sales = monthly_sales.max()

print(f"Highest Sales Month : {highest_month}")
print(f"Sales : {highest_month_sales:,.2f}")
```

```python
monthly_profit = (
    df.groupby("Month")["Profit"]
      .sum()
      .sort_index()
)

highest_profit_month = monthly_profit.idxmax()
highest_profit = monthly_profit.max()

print(f"Highest Profit Month : {highest_profit_month}")
print(f"Profit : {highest_profit:,.2f}")
```

```python
# Sales by State and Category
state_category_table = pd.pivot_table(
    df,
    values="TotalAmount",
    index="State",
    columns="Category",
    aggfunc="sum",
    fill_value=0
)

state_category_table
```

```python
sales_by_state.plot(
    kind="bar",
    figsize=(12,6),
    title="Sales by State"
)

plt.xlabel("State")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sales_by_category.plot(
    kind="pie",
    figsize=(8,8),
    autopct="%1.1f%%"
)

plt.title("Category-wise Sales")
plt.ylabel("")
plt.show()
```

```python
monthly_sales.plot(
    kind="line",
    marker="o",
    figsize=(12,6),
    title="Monthly Sales Trend"
)

plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)

plt.show()
```

```python
top10_products.plot(
    kind="barh",
    figsize=(10,6),
    title="Top 10 Products by Sales"
)

plt.xlabel("Sales")
plt.ylabel("Product")

plt.tight_layout()
plt.show()
```
